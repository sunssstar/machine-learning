```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df.head(3)
```

```python
df['label'] = iris.target
df.head(3)
```

```python
# x,y >> train / test data êµ¬ë¶„

x = iris.data
y = df['label']

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test =\
train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.shape, x_test.shape
```

```python
y_train.shape, y_test.shape
```

```python
# ë¨¸ì‹ ëŸ¬ë‹ ì•Œê³ ë¦¬ì¦˜ ì„ íƒ >> ëª¨ë¸ë§

# ì˜ì‚¬ê²°ì •ë‚˜ë¬´ (DecissionTreeClassifier)

from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()

# í›ˆë ¨(ë°˜ë“œì‹œ, í›ˆë ¨ìš© ë°ì´í„°) >> ëª¨ë¸ êµ¬ì¶•
dt.fit(x_train, y_train)
```

```python
# ëª¨ë¸ ì„±ëŠ¥í‰ê°€(ì •í™•ë„ í‰ê°€)

from sklearn.metrics import accuracy_score

# ì˜ˆì¸¡ (í‰ê°€ìš© ë°ì´í„°ë¡œ )
y_pred = dt.predict(x_test)
accuracy_score(y_test, y_pred)
```
