```python
import pandas as pd
import numpy as np

from sklearn import linear_model
from sklearn.metrics import mean_squared_error

from sklearn.model_selection import KFold
```

```python
# Ridge regression

reg = linear_model.Ridge(alpha = 1)
# alpha : tuning parameter (lambda 상수)
```

```python
# 훈련
reg.fit([[0,0],[0,0],[1,1]],[0,.1,1])
# x =[[0,0],[0,0],[1,1]]
# y = [0,.1,1]
```

```python
# test
pred_test = reg.predict([[0,1]])
pred_test
```

```python
# coefficient
reg.coef_
```

Advertisting 데이터 활용 모델 선택

```python
ad = pd.read_csv('/content/drive/MyDrive/MLP_bigdata/data/Advertising.csv',index_col = 0)
ad.head()
```

```python
# ad 데이터에 Lasso 적용 ==> 최적의 lambda 선정

n_fold = 5

for t_param in (0.001, 1 , 1000):
  # lambda 값 설정 0.001, 1 , 1000
  print( 'Lambda :',t_param)
  kf= KFold(n_splits = n_fold)
  idx =1

  sum_val_mse = 0
  for train, val in kf.split(ad):
    print('Fold: #', idx)

    # training set 의 feature와 response 분리
    train_x = ad.iloc[train][['TV','Radio','Newspaper']]
    train_y = ad.iloc[train][['Sales']]

    # test set 의 feature와 response 분리
    val_x = ad.iloc[val][['TV','Radio','Newspaper']]
    val_y = ad.iloc[val][['Sales']]

    # lasso 적용
    regr =linear_model.Lasso(alpha=t_param)

    # 훈련
    regr.fit(train_x[['TV','Radio','Newspaper']],train_y)

    #coef
    print('Coef :',regr.coef_)

    # validation set 예측
    val_y_pred = regr.predict(val_x[['TV','Radio','Newspaper']])

    # validation MSE
    val_mse = mean_squared_error(val_y, val_y_pred)

    sum_val_mse = val_mse

    print('--' * 100)
    idx +=1

  print('Average MSE :%.3f'% (sum_val_mse/n_fold))
  print('**'*100)
```
